```python
def calc_s_v(velocities_encper100ms, voltages):
    velocities = numpy.array(velocities_encper100ms)
    circumference = inch_to_meter(2.12) * numpy.pi * 2
    ticks_per_rev = 4096.
    velocities = velocities / ticks_per_rev * circumference * 10
    plt.plot(velocities, voltages, 'o')

    r = stats.linregress(velocities, voltages)
    voltages2 = velocities * r.slope + r.intercept
    plt.plot(velocities, voltages2)
    plt.show()
    print("Ks: ", r.intercept)
    print("Kv: ", r.slope)
    
print("raise, first stage")
calc_s_v([-750, -1050], [-8.15, -10.3])
print("raise, second stage")
calc_s_v([-520, -810], [-7.7, -9.65])
print("lower")
calc_s_v([1170, 1540], [8.76, 11.35])
```
