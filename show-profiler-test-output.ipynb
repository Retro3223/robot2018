```python
from frc3223_azurite.data import read_csv
from matplotlib import pyplot as plt
```

```python
data = read_csv("./logs/test_profiled_forward2.csv")
ts = data['t']
vs = data['v']
vmaxs = data['max_v']
amaxs = data['max_a']
xs = data['pos']
rs = data['target_pos']
accels = data['a']
voltages = data['vpl']
adist = data['adist']
err = data['err']
```

```python
s = 0
e = len(ts)//1
plt.plot(ts[s:e], vs[s:e])
plt.plot(ts[s:e], vmaxs[s:e])
plt.ylabel('velocity')
plt.show()
plt.plot(ts[s:e], accels[s:e])
plt.plot(ts[s:e], amaxs[s:e])
plt.ylabel('acceleration')
plt.show()
plt.plot(ts[s:e], xs[s:e])
plt.plot(ts[s:e], rs[s:e])
plt.ylabel('position')
plt.show()
plt.plot(ts[s:e], voltages[s:e])
plt.ylabel('voltage')
plt.show()
plt.plot(ts[s:e], adist[s:e])
plt.plot(ts[s:e], err[s:e])
plt.show()
```
