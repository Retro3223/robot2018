```python
from frc3223_azurite.data import read_csv
from matplotlib import pyplot as plt
```

```python
data = read_csv("./logs/test_StateSpaceDriveCommand.csv")
ts = data['t']
vrs = data['vel_r_mps']
vls = data['vel_l_mps']
xls = data['pos_l_m'] 
xrs = data['pos_r_m'] 
mxls = data['m_pos_l_m'] 
mxrs = data['m_pos_r_m'] 
target_xls = data['target_pos_l_m']
target_xrs = data['target_pos_r_m']
target_vls = data['target_vel_l_mps']
target_vrs = data['target_vel_r_mps']
Vs = data['voltage']
vpl = data['vpl']
vpr = data['vpr']
```

```python
s = 0
e = len(ts)//1
plt.plot(ts[s:e], vls[s:e])
plt.plot(ts[s:e], target_vls[s:e])

plt.ylabel('velocity left')
plt.legend(["actual", "target"])
plt.show()
plt.plot(ts[s:e], vrs[s:e])
plt.plot(ts[s:e], target_vrs[s:e])
plt.ylabel('velocity right')
plt.legend(["actual", "target"])
plt.show()
plt.plot(ts[s:e], xls[s:e])
plt.plot(ts[s:e], target_xls[s:e])
plt.plot(ts[s:e], mxls[s:e], 'o')
plt.ylabel('position left')
plt.legend(["actual", "target", 'measured'])
plt.show()
plt.plot(ts[s:e], xrs[s:e])
plt.plot(ts[s:e], target_xrs[s:e])
plt.plot(ts[s:e], mxrs[s:e], 'o')
plt.ylabel('position right')
plt.legend(["actual", "target", 'measured'])
plt.show()
plt.plot(ts[s:e], vpl[s:e])
plt.plot(ts[s:e], vpr[s:e])
plt.ylabel('voltage %')
plt.legend(["left", "right"])
plt.show()
```
